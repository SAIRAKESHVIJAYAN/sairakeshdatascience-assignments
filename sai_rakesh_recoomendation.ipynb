```python
import pandas as pd
df = pd.read_csv('anime.csv')
df.head()
```

```python
genres_dummies = df['genre'].str.get_dummies(sep=', ')
df = pd.concat([df, genres_dummies], axis=1)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df['normalized_rating'] = scaler.fit_transform(df[['rating']])
```

```python
from sklearn.metrics.pairwise import cosine_similarity

def recommend_anime(target_title, df, top_n=5):
    target_index = df[df['title'] == target_title].index[0]

    feature_columns = ['normalized_rating'] + list(genres_dummies.columns)
    cosine_matrix = cosine_similarity(df[feature_columns])

    similarity_scores = list(enumerate(cosine_matrix[target_index]))


    sorted_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)[1:top_n+1]
    recommended_indices = [i[0] for i in sorted_scores]

    return df['title'].iloc[recommended_indices]
```

```python
from sklearn.model_selection import train_test_split

train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)
```

Can you explain the difference between user-based and item-based collaborative filtering?

User-Based Collaborative Filtering: This approach recommends items by finding similar users based on their preferences. For instance, if User A and User B rate similar items highly, the system will recommend items rated highly by User A to User B.
Item-Based Collaborative Filtering: This method focuses on finding similarities between items rather than users. It recommends items based on the ratings of similar items. For example, if a user likes Anime A, the system will recommend Anime B because they share similar ratings from other users.
What is collaborative filtering, and how does it work?

Collaborative Filtering is a recommendation technique that relies on user-item interactions to predict a user’s preference for an item. It works by:
Analyzing historical data of users and their interactions with items.
Finding patterns in user behavior or item ratings.
Making recommendations based on similarities either between users or between items.
The fundamental idea is that users who agreed in the past will likely agree in the future regarding other items.
